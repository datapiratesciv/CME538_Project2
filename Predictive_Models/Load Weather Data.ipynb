```python
# Import 3rd party libraries
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', None)
```

```python
trips_filenames = [filename for filename in os.listdir()]
weather_data = pd.DataFrame()
for file in trips_filenames:
    if("climate" not in file):
      continue

    print(file)  
    temp_data = pd.read_csv(file)

    # change "Start Time" to a datetime object
    print(temp_data.head(5))
    for col_name in ["Date/Time", "Date/Time (LST)"]:
        try:
          temp_data["Date/Time"] = pd.DatetimeIndex(temp_data[col_name]).tz_localize("EST")
          break
        except:
          continue

    weather_data = pd.concat([weather_data,temp_data])
    print(weather_data.tail())
```

```python
safe_data = weather_data.copy().sort_values(by='Date/Time').copy()
#print(safe_data)
```
